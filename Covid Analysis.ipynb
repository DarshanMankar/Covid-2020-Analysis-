## Exploratory Data Analysis

This is simple EDA example for beginner to play with data.

```python
import numpy as np              # used to read and preprocess data
import seaborn as sns           # used to working with arrays Single or MultiDiementional
import pandas as pd             # Visualisation of data
import matplotlib.pyplot as plt # Visualisation of data
%matplotlib inline 
```

```python
# we are adding json file from Ministry of health and family welfare India portal
df=pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
```

```python
df
```

```python
# we observed that row 36 is not required for EDA so we eliminate this row.
# and also we dont want sr.no column.
df=df.iloc[:-1,1:]
```

```python
df
```

```python
df.head()
```

```python
# finding total no of states and stored into total_state array so we can draw graps wrt states.
total_state=np.arange(len(df['state_name']))
```

```python
total_state
```

```python
df['active']
```

### Total Active cases in each state of India

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(10,9),dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_state,df['positive'],align='center',alpha=0.8,color=(1,0,0),edgecolor=(0.5,0.6,0.2))
plt.yticks(total_state,df['state_name'])
plt.xlim(1,max(df['positive'])+10)                # limit of x axis will be total positive count + 10
plt.xlabel('Number of Active Cases')
plt.ylabel('States')
plt.title('Total Active Corona Cases')
```

### Total Active cases in each state of India

```python
x = df.groupby('state_name')['active'].mean().sort_values()
x.plot(kind='barh',figsize=(10,10), color='#292734', zorder=2, width=0.85)
plt.xlabel('Number of Active Cases')
plt.ylabel('States')
plt.title('Total Active Corona Cases')
```

### Total Positive cases in each state of India

```python
x = df.groupby('state_name')['positive'].mean().sort_values()
x.plot(kind='barh',figsize=(10,10), color='#86bf91', zorder=2, width=0.85)
plt.xlabel('Number of Positive Cases')
plt.ylabel('States')
plt.title('Total Positive Corona Cases')
```

### Total Cured cases in each state of India
